```python
for i in range(4,10):
    x= browser.find_element(By.XPATH,'//tr['+str(i)+']/td')
    print (i,x.text)
    rejected = False
    docNum = x.text
    if x.text=='NONE':
        rejected=True
    time.sleep(1)
    if not rejected:
        x.click()
        time.sleep(1)
        browser.find_element(By.XPATH,'//a[@alt="View in Patent Full Text and Image Database (PatFT)"]').click()


    else:
        x=browser.find_element(By.XPATH,'//tr['+str(i)+']/td[2]')
        docNum=x.text
        print (docNum)
        x.click()
        time.sleep(1)
        browser.find_element(By.XPATH,'//a[@alt="View in Patent Application Full Text and Image Database (AppFT)"]').click()
    browser.switch_to_window(browser.window_handles[1])
    name = None
    if rejected:
        name = str(docNum) + '_1.txt'
    else:
        name = str(docNum) +'_0.txt'
    browser.find_element_by_tag_name('body').text
    with open(name,'w') as file:
        file.write(browser.find_element_by_tag_name('body').text)
    browser.quit()
    browser = initialize(svc)
    browser.get('https://assignment.uspto.gov/patent/index.html#/patent/search/result?id='+Compnay+'&type=patAssigneeName')
    time.sleep(1)
    

    
```
